```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
# data
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
xx, yy = np.meshgrid(x, y, sparse=True)
```

```python
# kernel
delta = 1
l = 10 
a = -(((xx - yy) / l) ** 2) / 2
kernel = delta**2 * np.exp(a)
```

```python
# 绘图
fig = plt.figure(figsize=(12, 10))

im = plt.contourf(x, y, kernel)

# 添加颜色条
plt.colorbar(im)

# 添加标题和标签
plt.title("Heatmap Example")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# 显示图形
plt.show()
```

```python
kernel2 = 1 / 2 * xx * yy
# 绘图
fig = plt.figure(figsize=(12, 10))

im = plt.contourf(x, y, kernel2)

# 添加颜色条
plt.colorbar(im)

# 添加标题和标签
plt.title("Heatmap Example")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# 显示图形
plt.show()
```

```python
l = 1
p = 4
kernel3 = np.exp(2/l**2*np.sin(np.pi/p*np.abs(xx-yy))**2)
# 绘图
fig = plt.figure(figsize=(12, 10))

im = plt.contourf(x, y, kernel3)

# 添加颜色条
plt.colorbar(im)

# 添加标题和标签
plt.title("Heatmap Example")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# 显示图形
plt.show()
```
