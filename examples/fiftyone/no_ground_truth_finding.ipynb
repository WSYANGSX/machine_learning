```python
import fiftyone as fo
import fiftyone.zoo as foz

# 加载验证集
dataset = foz.load_zoo_dataset("coco-2017")
val_view = dataset.match_tags("validation")
val_view
```

```python
val_view.first()
```

```python
empty_count = 0
for sample in val_view:
    if not hasattr(sample.ground_truth, "detections"):
        empty_count+=1
        print(sample.id)

print(f"验证集中无标注的图像数量：{empty_count}")
```

```python
train_view = dataset.match_tags("train")

empty_count = 0
for sample in train_view:
    if not hasattr(sample.ground_truth, "detections"):
        empty_count += 1
        print(sample.id)

print(f"验证集中无标注的图像数量：{empty_count}")
```
